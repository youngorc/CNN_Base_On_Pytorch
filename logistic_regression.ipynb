```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
```

```python
input_size=28*28
num_classes=10
num_epochs=10
batch_size=64
learning_rate=0.001

train_dataset=torchvision.datasets.MNIST("../../data",train=True,transform=transforms.ToTensor(),download=False)
test_dataset=torchvision.datasets.MNIST("../../data",train=False,transform=transforms.ToTensor(),download=False)
train_loader=torch.utils.data.DataLoader(train_dataset,batch_size=batch_size,shuffle=True,drop_last=True)
test_loader=torch.utils.data.DataLoader(test_dataset,batch_size=batch_size,shuffle=False)
```

```python
x,y=train_dataset[0]
```

```python
model=nn.Linear(input_size,num_classes)
optim=torch.optim.SGD(model.parameters(),lr=learning_rate)
loss_func=nn.CrossEntropyLoss()

for epoch in range(num_epochs):
    for x,y in train_loader:
        x_train=x.view(-1,28*28)
        y_hat=model(x_train)
        loss=loss_func(y_hat,y)
        optim.zero_grad()
        loss.backward()
        optim.step()
    print("Epoch:{}/{},\tloss:{:.4f}".format(epoch+1,num_epochs,loss.item()))
```

```python
with torch.no_grad():
    correct=0
    total=0
    for i,test_data in enumerate(test_loader):
        images,labels=test_data
        images=images.view(-1,28*28)
        outputs=model(images)
        predicted=torch.argmax(outputs,axis=1)
        total+=outputs.size(0)
        correct+=(predicted==labels).sum()
        print("{} time,Accuracy:{}/{}".format(i,correct,total))
```
